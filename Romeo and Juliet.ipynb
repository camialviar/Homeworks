```python
%matplotlib inline 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

```python
# 2d Linear system
def ode(state,t):
    R = state[0] # Romeo's love for Juliet. Negative values indicate hate.
    J = state[1] # Juliet's love for Romeo. Negative values indicate hate.
    
    
     
    # System parameters. Play with these to see different dynamics!!
    
    a = -.4 # Romeo's awareness of his own feelings.
    b = -.3 # Romeo's beliefs about Juliet's feelings for him. Positive values: Romeo's beliefs are in accordance with 
              # Juliet's feelings. Negative values: Romeo's beliefs about Juliet's feelings are contrary to Juliet's 
              # feelings. 
    
    c = -.1 # Juliet's awareness of her own feelings.
    d = -.1 # Juliet's beliefs about Romeo's feelings for her. Positive values: Juliet's beliefs are in accordance with 
              # Romeo's feelings. Negative values: Juliet's beliefs about Romeo's feelings are contrary to Romeo's 
              # feelings. 
    
    # Main equation here
    dR = a*R + b*J # Each of the lovers' feelings for the other depend upon their awareness of their previous 
                   # feelings and their beliefs about their lover's feelings. 
    dJ = c*J + d*R
    
    # Return the state derivatives
    return [dR,dJ]
```

```python
t = np.arange(0, 10, .01) # Times

# A grid of initial conditions to test
for R,J in itertools.product(range(-5,5), range(-2,2)):
    ic = [R,J] # initial condition
    orbit = odeint(ode, ic, t) # Run the dst; integrate the ode
    plt.plot(ic[0],ic[1], '.', color = 'k')
    plt.plot(orbit[:,0],orbit[:,1], color='blue') #Arrows?
```
